```python
# Step 1 - Import Libraries

import numpy as np
import pandas as pd
```

```python
# Step 2 - Load Dataset

def load_fitness_data(file_path):
    try:
        # Read CSV file
        df = pd.read_csv(file_path)

        # Store Date column separately
        dates = df["Date"].tolist()

        # Convert numeric columns to NumPy array
        data = df[["Steps", "SleepHours", "CaloriesBurned"]].to_numpy()

        print("Dataset Loaded Successfully!")
        print("Shape of Dataset:", data.shape)

        return dates, data

    except Exception as e:
        print("Error loading dataset:", e)
        return None, None
```

```python
# Dataset URL
file_path = "https://samatrix-data.s3.ap-south-1.amazonaws.com/Statistics-Project/fitness_data.csv"

# Load Data
dates, data = load_fitness_data(file_path)

# Display first 5 dates
print(dates[:5])

# Display first 5 rows
print(data[:5])
```

```python
# Step 3 - Weekly Analysis

def weekly_analysis(data, week_number):
    start = (week_number - 1) * 7
    end = start + 7

    week_data = data[start:end]

    avg = np.mean(week_data, axis=0)

    print(f"\n===== Week {week_number} =====")
    print(f"Average Steps       : {avg[0]:.2f}")
    print(f"Average Sleep Hours : {avg[1]:.2f}")
    print(f"Average Calories    : {avg[2]:.2f}")

    return week_data
```

```python
# Run Weekly Analysis

for week in range(1, 5):
    weekly_analysis(data, week)
```

```python
# Step 4 - Find Low Activity Days

def find_low_activity_days(data, dates):

    print("\n===== Low Activity Days =====")

    low_steps = data[:, 0] < 6000
    low_sleep = data[:, 1] < 6
    low_calories = data[:, 2] < 1800

    found = False

    for i in range(len(data)):
        if low_steps[i] or low_sleep[i] or low_calories[i]:
            found = True
            print(f"\nDate: {dates[i]}")
            print(f"Steps: {data[i,0]}")
            print(f"Sleep Hours: {data[i,1]}")
            print(f"Calories Burned: {data[i,2]}")

    if not found:
        print("No low activity days found.")
```

```python
find_low_activity_days(data, dates)
```

```python
# Step 5 - Daily Fitness Score

def fitness_score(day):

    steps, sleep, calories = day

    step_score = min(steps / 10000, 1)
    sleep_score = min(sleep / 8, 1)
    calorie_score = min(calories / 2200, 1)

    score = (0.4 * step_score) + (0.3 * sleep_score) + (0.3 * calorie_score)

    return round(score * 100, 2)
```

```python
print("Fitness Score of Day 1:", fitness_score(data[0]))
```

```python
# Step 6 - Score Every Day

def score_all_days(data):
    scores = []

    for day in data:
        scores.append(fitness_score(day))

    return np.array(scores)
```

```python
scores = score_all_days(data)

print("All Fitness Scores:")
print(scores)
```

```python
# Step 7 - Monthly Summary

def monthly_summary(data, dates):

    print("\n========== MONTHLY SUMMARY ==========")

    best_day = np.argmax(data[:,0])
    worst_day = np.argmin(data[:,0])

    print(f"\nBest Step Day : {dates[best_day]}")
    print(f"Steps : {data[best_day,0]}")

    print(f"\nWorst Step Day : {dates[worst_day]}")
    print(f"Steps : {data[worst_day,0]}")

    scores = score_all_days(data)

    top3 = np.argsort(scores)[-3:][::-1]

    print("\nTop 3 Fitness Days")

    for i in top3:
        print(f"{dates[i]} ---> Score : {scores[i]}")
```

```python
monthly_summary(data, dates)
```

```python
# Step 8 - Save Summary Report

def save_summary(data):

    active_days = np.sum(data[:,0] >= 10000)

    avg_sleep = np.mean(data[:,1])

    avg_calories = np.mean(data[:,2])

    with open("fitness_summary.txt","w") as file:

        file.write("FITNESS TRACKER SUMMARY\n")
        file.write("=======================\n\n")

        file.write(f"Active Days : {active_days}\n")
        file.write(f"Average Sleep : {avg_sleep:.2f} Hours\n")
        file.write(f"Average Calories Burned : {avg_calories:.2f}\n")

    print("\nfitness_summary.txt saved successfully.")
```

```python
save_summary(data)
```

```python
print("\n========= FITNESS TRACKER PROJECT =========")

if data is not None:

    for week in range(1,5):
        weekly_analysis(data, week)

    find_low_activity_days(data, dates)

    monthly_summary(data, dates)

    save_summary(data)

else:
    print("Dataset could not be loaded.")
```
